## Disease Prediction from Symptoms

```python
# Import Dependencies
import csv                            #melakukan import module csv
import pandas as pd                   #melakukan import module pandas untuk kebutuhan table
import numpy as np                    #melakukan import module numpy untuk array dan matriks
from collections import defaultdict   #melakukan import kelas defauldict dari module collection
import seaborn as sns                 #melakukan import module seaborn untuk kebutuhan visualisasi
import matplotlib.pyplot as plt       #melakukan import matplotlib untuk kebuuhan visualisasi
%matplotlib inline                    #melakukan pengaktifan matplotlib inline untuk kebutuhan menampilkan plot secara otomatis
%matplotlib inline                    #melakukan pengaktifan matplotlib inline untuk kebutuhan menampilkan plot secara otomatis
```

```python
# Fill all NaN with the values above
data = df.fillna(method='ffill')       # untuk mengisi data yang kosong atau hilang (NaN)
```

```python
data.head()                            # untuk menampilkan 5 baris pertama dari data yang dimulai dari angka 0
```

```python
# Process Disease and Symptom Names
def process_data(data):                              #melakukan proses "data" nama penyakit dan gejala
    data_list = []                                   #membuat tempat untuk menampung data array
    data_name = data.replace('^','_').split('_')     #proses merubah tanda dari data
    n = 1
    for names in data_name:
        if (n % 2 == 0):
            data_list.append(names)
        n += 1
    return data_list                                 #menreturn nilai data_list
```

```python
# Data Cleanup
disease_list = []
disease_symptom_dict = defaultdict(list)            #menampung data penyakit
disease_symptom_count = {}                          #menampung data list
count = 0                                           #menampung gejala

for idx, row in data.iterrows():                    #melakukan loop data


    # Get the Disease Names
    if (row['Disease'] !="\xc2\xa0") and (row['Disease'] != ""):   #jika row['Disease'] !="\xc2\xa0" dan nilai kolom disease tidak kosong
        disease = row['Disease']
        disease_list = process_data(data=disease)
        count = row['Count of Disease Occurrence']

    # Get the Symptoms Corresponding to Diseases
    if (row['Symptom'] !="\xc2\xa0") and (row['Symptom'] != ""):   #jika row['symptom'] dan nilai kolom symptom tidak kosong
        symptom = row['Symptom']
        symptom_list = process_data(data=symptom)
        for d in disease_list:
            for s in symptom_list:
                disease_symptom_dict[d].append(s)
            disease_symptom_count[d] = count
```

```python
# See that the data is Processed Correctly  #untuk memeriksa data apakah diproses dengan benar
disease_symptom_dict
```

```python
# Count of Disease Occurence w.r.t each Disease
disease_symptom_count                            #menghitung kemunculan penyakit
```

```python
# Save cleaned data as CSV
f = open('/content/cleaned_data.csv', 'w')     #membuat file baru untuk menyimpan data cleaned_data.csv


with f:
    writer = csv.writer(f)
    for key, val in disease_symptom_dict.items():
        for i in range(len(val)):
            writer.writerow([key, val[i], disease_symptom_count[key]])
```

```python
# Read Cleaned Data as DF
df = pd.read_csv('/content/cleaned_data.csv')           #membaca data csv cleaned_data.csv
df.columns = ['disease', 'symptom', 'occurence_count']  #mengubah nama kolom
df.head()                                               #menampilkan hasil kolom
```

```python
# Remove any rows with empty values
df.replace(float('nan'), np.nan, inplace=True)   #mengganti nilai yang kosong/Nan
df.dropna(inplace=True)                          #menghapus baris yang mengandung nilai kosong/Nan
```

```python
# Encode the Labels
from sklearn.preprocessing import LabelEncoder                #mengimport modul LabelEncoder dari sklearn.preprocessing
from sklearn.preprocessing import OneHotEncoder               #mengimport OneHotEncoder dari sklearn.preprocessing

label_encoder = LabelEncoder()                                #membuat sebuah objek LabelEncoder untuk menampung data
integer_encoded = label_encoder.fit_transform(df['symptom'])  #memberikan kode untuk nilai data symptom
print(integer_encoded)                                        #mencetak hasil
```

```python
# One Hot Encode the Labels
onehot_encoder = OneHotEncoder(sparse=False)                         #Menginisialisasi objek OneHotEncoder
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)   #Mereshape array input integer_encoded agar memiliki satu kolom menggunakan .reshape(len(integer_encoded) menjadi 2D
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)       #melakukan one-hot encoding menggunakan metode fit_transform() dari objek OneHotEncoder
print(onehot_encoded)                                                #mencetak hasil
```

```python
cols = np.asarray(df['symptom'].unique())    #mengambil nilai unik dari kolom 'symptom' dalam dataframe df dan menyimpannya dalam array cols
cols                                         #berisi daftar nilai unik dari kolom 'symptom'.
```

```python
# Create a new dataframe to save OHE labels
df_ohe = pd.DataFrame(columns = cols)        #membuat dataframe dengan kolom-kolom yang sesuai dengan nilai unik dari kolom 'symptom'
df_ohe.head()                                #mencetak dataframe kosong dengan kolom-kolom yang telah ditentukan sesuai dengan nilai unik dari kolom 'symptom
```

```python
for i in range(len(onehot_encoded)):       #loop untuk memasukan setiap baris dari array onehot_encoded ke dalam dataframe df_ohe
    df_ohe.loc[i] = onehot_encoded[i]
```

```python
# Disease Dataframe
df_disease = df['disease']   #membuat DataFrame df_disease yang berisi kolom 'disease' dari DataFrame df. df['disease']
df_disease.head()            #menampilkan beberapa baris pertama dari kolom 'disease'
```

```python
# Concatenate OHE Labels with the Disease Column
df_concat = pd.concat([df_disease,df_ohe], axis=1)  #menggabungkan kolom 'disease' dari DataFrame df_disease dengan DataFrame df_ohe, yang berisi hasil one-hot encoding dari kolom 'symptom
df_concat.head()                                    #mendapatkan DataFrame df_concat yang berisi kolom 'disease' dan hasil one-hot encoding dari kolom 'symptom', yang telah digabungkan bersama
```

```python
df_concat.drop_duplicates(keep='first',inplace=True)    #menghapus baris-baris duplikat dari DataFrame df_concat
```

```python
cols = cols[1:]             # mendapatkan subset dari array cols, dimulai dari indeks ke-1 hingga akhir
```

```python
# Since, every disease has multiple symptoms, combine all symptoms per disease per row
df_concat = df_concat.groupby('disease').sum()  #mengelompokkan berdasarkan kolom 'disease' dan jumlahkan nilai gejalanya
df_concat = df_concat.reset_index()             #mengembalikan indeks DataFrame ke bentuk awalnya setelah pengelompokan
df_concat[:5]                                   #menampilkan beberapa baris pertama dari DataFrame yang sudah dimodifikasi.
```

```python
# One Hot Encoded Features
X = df_concat[cols]   #menggunakan kolom-kolom one-hot encoded (disebut cols) dari DataFrame df_concat

# Labels
y = df_concat['disease']   #memprediksi penyakit berdasarkan gejala-gejala yang diberikan
```

## Model Training

```python
from sklearn.model_selection import train_test_split                ##import module train_test split dari sckitlearn untuk membagi dataset
from sklearn.naive_bayes import MultinomialNB                       #import MultinomialNB dari naive bayes
from sklearn import tree                                            #import module tree dari decision tree classifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz    #import DecisionTreeClassifier untuk membuat model tree, export_graphviz untuk menyimpan visualisasi
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)      #membagi dataset test dan train dengan test_size 0,2 dan random_state 101
```

```python
dt = DecisionTreeClassifier()        #membuat kelas dt untuk DecisionTreeClassifier
clf_dt=dt.fit(X, y)                  #melatih model x dan y
```
